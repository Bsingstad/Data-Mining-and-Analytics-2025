```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
my = 0 # gjennomsnitt
sigma = 0.5 # standardavvik

norm_dist = np.random.normal(my, sigma, 100000)
sns.histplot(data=norm_dist, bins=20)
plt.show()
```

```python
pop_mean = norm_dist.mean()
```

```python
pop_std = norm_dist.std()
```

```python
delta_mean = []
delta_std = []

for n in np.linspace(10,5000,50):
    samples = np.random.choice(norm_dist, int(n))
    print("Antall samples: ", int(n))
    print("Gjennomsnitt: ", samples.mean())
    print("Standardavvik:", samples.std())
    print("Er normalfordelt: ", stats.normaltest(samples))
    delta_mean.append(samples.mean()- pop_mean)
    delta_std.append(samples.std() - pop_std)

    sns.histplot(data=samples, kde=True, bins=20)

    plt.show()
```

```python
plt.plot(delta_std, label="forskjell i standardavvik")
plt.plot(delta_mean, label="forskjell i gjennomsnitt")
plt.hlines(xmin=0,xmax=50, y=0, linestyle='--', colors="k")
plt.legend()
plt.show()
```
