## Oppgave 1
Sammenlign karakterer for skoleklasse A og B

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
klasse_A = np.random.randint(1, 7, size=30)
klasse_B = np.random.randint(1, 7, size=30)

print("Karakterer i A-klassen: ", klasse_A)
print("Karakterer i B-klassen: ", klasse_B)
plt.hist(klasse_A, alpha=0.5,label="A klassen")
plt.hist(klasse_B, alpha=0.5, label="B klassen")
plt.legend()
plt.show()
```

### Oppgave 5
Undersøk om det er en signifikant forskjell mellom jenter og gutter når det gjelder å bestå en bestemt eksamen 


```python
n_gutter = 40
n_jenter = 38
jenter = np.random.choice(['Bestått','Ikke_bestått'], size=n_jenter, p=[0.7,0.3])
gutter = np.random.choice(['Bestått','Ikke_bestått'], size=n_gutter, p=[0.5,0.5])

df = pd.DataFrame({'Klasser':['jenter']*n_jenter + ['gutter']*n_gutter, 'Resultat': np.concatenate([jenter, gutter])})
krysstabell = pd.crosstab(df['Klasser'], df['Resultat'])[['Bestått','Ikke_bestått']]
print(krysstabell)
```
