```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy.stats import mannwhitneyu
```

```python
elementer = [1,2,3,4]
sannsynlighet_gruppe_1 = [0.2, 0.5, 0.1, 0.2]
sannsynlighet_gruppe_2 = [0.1, 0.2, 0.2, 0.5]
gruppe_1 = np.random.choice(elementer, 100, p=sannsynlighet_gruppe_1)
gruppe_2 = np.random.choice(elementer, 100, p=sannsynlighet_gruppe_2)
```

```python
sns.histplot(data=gruppe_1, bins= 4, color="C0", label="gruppe1")
sns.histplot(data=gruppe_2,  bins=4, color="C1", label="gruppe2")
plt.xticks([1,2,3,4])
plt.legend()
plt.show()
```

```python
mannwhitneyu(gruppe_1, gruppe_2, method="exact")
```
