```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import  train_test_split 
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score
from scipy import stats
```

```python
# Last inn dataene
cancer = load_breast_cancer(as_frame=True)
```

```python
X_data = cancer["data"]
X_data.head()
```

```python
y_data = cancer["target"]
y_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data,y_data,random_state=42)
```

```python
from sklearn.model_selection import RepeatedKFold, GridSearchCV
```

```python
rkf = RepeatedKFold(n_splits=5, n_repeats=3, random_state=42)
```

```python
param_grid_log_reg =  {"penalty":("l1","l2","elasticnet", "None"), 
                       "C":[0.5, 1 ,2], 
                       "solver":["lbfgs", "liblinear"],
                       "max_iter":[1000, 2500,5000]}

```

```python
logreg = LogisticRegression()
```

```python
search_lr = GridSearchCV(estimator=logreg, param_grid=param_grid_log_reg, scoring="f1", cv=rkf)
search_lr.fit(X_train, y_train)
```

```python
param_grid_mlp =  {"alpha":[0.0001,0.001,0.01],  
                    "solver":["lbfgs", "sgd", "adam"],
                       "max_iter":[2500,5000,7500]}
```

```python
mlp = MLPClassifier()
```

```python
search_mlp = GridSearchCV(estimator=mlp, param_grid=param_grid_mlp, scoring="f1", cv=rkf)
search_mlp.fit(X_train, y_train)
```

```python
results_df_lr = pd.DataFrame(search_lr.cv_results_)
results_df_lr = results_df_lr.sort_values(by=["rank_test_score"])
results_df_lr = results_df_lr.set_index(
    results_df_lr["params"].apply(lambda x: "_".join(str(val) for val in x.values()))
).rename_axis("kernel")
results_df_lr[["params", "rank_test_score", "mean_test_score", "std_test_score"]].head()
```

```python
results_df_mlp = pd.DataFrame(search_mlp.cv_results_)
results_df_mlp = results_df_mlp.sort_values(by=["rank_test_score"])
results_df_mlp = results_df_mlp.set_index(
    results_df_mlp["params"].apply(lambda x: "_".join(str(val) for val in x.values()))
).rename_axis("kernel")
results_df_mlp[["params", "rank_test_score", "mean_test_score", "std_test_score"]].head()
```

```python
results_df_lr.params[0]
```

```python
rkf = RepeatedKFold(n_splits=5, n_repeats=10, random_state=42)
```

```python
f1_score_log_reg = []
for i, (train_index, val_index) in enumerate(rkf.split(X_train)):
    
    X_train_cv, X_val_cv = X_train.iloc[train_index], X_train.iloc[val_index]
    y_train_cv, y_val_cv = y_train.iloc[train_index], y_train.iloc[val_index]
    
    logreg = LogisticRegression(**results_df_lr.params[0])
    logreg.fit(X_train_cv, y_train_cv)
    
    y_val_hat = logreg.predict_proba(X_val_cv)[:,1]
    score = f1_score(y_val_cv,(y_val_hat>0.5)*1,average='binary').round(2)
    #print("F1-score = ",score)
    f1_score_log_reg.append(score)
```

```python
f1_score_mlp = []
for i, (train_index, val_index) in enumerate(rkf.split(X_train)):    
    X_train_cv, X_val_cv = X_train.iloc[train_index], X_train.iloc[val_index]
    y_train_cv, y_val_cv = y_train.iloc[train_index], y_train.iloc[val_index]
    
    mlp = MLPClassifier(**results_df_mlp.params[0])
    mlp.fit(X_train_cv, y_train_cv)
    
    y_val_hat = mlp.predict_proba(X_val_cv)[:,1]
    score = f1_score(y_val_cv,(y_val_hat>0.5)*1,average='binary').round(2)
    #print("F1-score = ",score)
    f1_score_mlp.append(score)
```

```python
log_reg_mean = np.asarray(f1_score_log_reg).mean().round(3)
log_reg_std = np.asarray(f1_score_log_reg).std().round(3)
print(r"Logistic Regression = {} Â± {}".format(log_reg_mean,log_reg_std))
```

```python
mlp_mean = np.asarray(f1_score_mlp).mean().round(3)
mlp_std = np.asarray(f1_score_mlp).std().round(3)
print(r"Logistic Regression = {} Â± {}".format(mlp_mean,mlp_std))
```

```python
plt.boxplot([np.asarray(f1_score_log_reg),np.asarray(f1_score_mlp)])
plt.ylim(0.8,1.05)
plt.xticks([1,2],["Log Reg","MLP"])
plt.ylabel("F1-score")
plt.xlabel("Modeller")
plt.grid()
plt.show()
```

```python
plt.hist(np.asarray(f1_score_log_reg),  label="Logistisk regresjon")
plt.hist(np.asarray(f1_score_mlp), alpha=0.7,bins=40, label="Nevralt nettverk")
plt.xlim(0.8,1.0)
plt.xlabel("F1-score")
plt.ylabel("Antall")
plt.legend()
plt.show()
```

```python
print("p-verdi = ",stats.ttest_ind(np.asarray(f1_score_log_reg), np.asarray(f1_score_mlp), axis=0,equal_var=True, alternative="two-sided")[1].round(7))
```
